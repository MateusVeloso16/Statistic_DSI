Gerador de Numeros Aleatorios

```python
import numpy as np

rng = np.random.default_rng()
print(rng)
```

```python
numero_aleatorio = rng.random() * 10
print(f'{numero_aleatorio:.3f}')
```

```python
array_aleatorio = rng.random(5)
print(array_aleatorio)
```

```python
rng = np.random.default_rng(seed=42)
dados_vendas = rng.integers(low=50, high=200, size=30)
print(dados_vendas)
```

```python
print(np.max(dados_vendas))
```

```python
print(np.argmax(dados_vendas) + 1)
```

```python
print(f'Dia com mais vendas: {np.argmax(dados_vendas) + 1}.')
```

```python
print(f'Dia com menos vendas: {np.argmin(dados_vendas) + 1}.')
```

```python
media_vendas = np.mean(dados_vendas)
print(f'A media de vendas durante o mes foi de {int(media_vendas)} vendas por dia.')
```

```python
import numpy as np

tempos_ciclo = np.array([5.5, 5.7, 5.9, 6.0, 5.8, 5.6, 5.7, 7.2, 4.8])
media = np.mean(tempos_ciclo)
print(media)
```

```python
desvio_padrao = np.std(tempos_ciclo)
print(desvio_padrao)
```

```python
condicao = tempos_ciclo > 5
print(np.where(condicao))
```

```python
condicao = (tempos_ciclo > media + 2 * desvio_padrao) | (tempos_ciclo < media - 2 * desvio_padrao)
anomalias = np.where(condicao)
print(anomalias)
```

```python
print(tempos_ciclo[anomalias])
```

```python
print(tempos_ciclo[condicao])
```

```python
print(f'Os tempos que estao na condicao pedida sao {tempos_ciclo[condicao]}.')
```
