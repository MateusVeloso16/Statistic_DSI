```python
import numpy as np
import pandas as pd

```

```python
X = np.array([[1.2, 3], 
              [1.5, 4], 
              [1.0, 2], 
              [2.0, 5], 
              [1.8, 4]])

print("X:")
print(X)
```

```python
W = np.array([[80],   # Weight for size
              [20]])  # Weight for number of rooms

print("\nW (in €000):")
print(W)
```

```python
y_pred = X @ W  # or np.dot(X, W)
print("\nPredicted Price (X @ W):")
print(y_pred)
```

```python
Y = np.array([[151],
              [204],
              [117],
              [266],
              [222]])

print("\nActual Prices (Y):")
print(Y)
print("\nShape of Y:", Y.shape)
```

```python
error = Y - y_pred

print("Error:")
print(error)
```

```python
SSE = error.T @ error

print("\nSSE")
print(SSE)
```

```python
W_new = np.array([[81],
                    [19]])

y_pred_new = X @ W_new
error_new = Y - y_pred_new
SSE_new = error_new.T @ error_new
SSE_new
```
